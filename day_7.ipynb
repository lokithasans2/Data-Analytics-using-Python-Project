```python
import pandas as pd
data = {
    "Region": ["South", "North", "East", "West", "South", "North"],
    "Product": ["Laptop", "Mobile", "Laptop", "Tablet", "Mobile", "Tablet"],
    "Sales": [50000, 30000, 45000, 25000, 20000, 35800]
}
df = pd.DataFrame(data)
print(df)
```

```python
south_sales = df[df['Region'] == 'South']
print(south_sales)
```

```python
sorted_df = df.sort_values("Sales")
print(sorted_df)
```

```python
import pandas as pd
data = {
    "Region": ["South", "North", "East", "West", "South", "North"],
    "Product": ["Laptop", "Mobile", "Laptop", "Tablet", "Mobile", "Tablet"],
    "Sales": [50000, 30000, 45000, 25000, 20000, 35800]
}
df = pd.DataFrame(data)
print(df)
```

```python
region_sales = df.groupby("Region") ["Sales"].sum()

print(region_sales)
```

```python
Product_sales = df.groupby("Product") ["Sales"].sum()

print(Product_sales)
```

```python
import pandas as pd
data = {
    "Region": ["South", "North", "East", "West", "South", "North"],
    "Product": ["Laptop", "Mobile", "Laptop", "Tablet", "Mobile", "Tablet"],
    "Sales": [50000, 30000, 45000, 25000, 20000, 35800]
}
df = pd.DataFrame(data)
print(df)
```

```python
result = df.groupby("Region") ["Sales"].agg(
    ["sum", "mean", "max", "min"]
)
print(result)
```
