```python
# Student Performance & Result Analytics

import pandas as pd
import matplotlib.pyplot as plt

# Student Marks Data
data = {
    "Student": ["Suruthis", "Saifulla", "Manoj", "Sakthi", "Adhil"],
    "Maths": [85, 65, 40, 95, 55],
    "Science": [78, 70, 35, 92, 60],
    "English": [88, 75, 45, 90, 58]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate Average Marks
df["Average"] = (df["Maths"] + df["Science"] + df["English"]) / 3

# Calculate Grade
def calculate_grade(avg):
    if avg >= 90:
        return "A+"
    elif avg >= 80:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 60:
        return "C"
    elif avg >= 50:
        return "D"
    else:
        return "F"

df["Grade"] = df["Average"].apply(calculate_grade)

# Pass / Fail Status
df["Status"] = df["Average"].apply(
    lambda x: "Pass" if x >= 50 else "Fail"
)

# Display Student Report
print("\n===== STUDENT PERFORMANCE REPORT =====")
print(df)

# Find Topper
topper = df.loc[df["Average"].idxmax()]
print("\nTopper:")
print(topper)

# Find Weak Student
weak_student = df.loc[df["Average"].idxmin()]
print("\nWeak Student:")
print(weak_student)

# Subject-wise Performance
subject_avg = df[["Maths", "Science", "English"]].mean()

print("\nSubject-wise Average Marks:")
print(subject_avg)

# Chart 1: Subject-wise Average

plt.figure(figsize=(6,4))
plt.bar(subject_avg.index, subject_avg.values)
plt.title("Subject-wise Average Marks")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()

# Chart 2: Student Average Marks

plt.figure(figsize=(6,4))
plt.bar(df["Student"], df["Average"])
plt.title("Student Average Marks")
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.show()

# Chart 3: Pass/Fail Analysis

status_count = df["Status"].value_counts()

plt.figure(figsize=(5,5))
plt.pie(status_count,
        labels=status_count.index,
        autopct="%1.1f%%")
plt.title("Pass vs Fail Percentage")
plt.show()
```
