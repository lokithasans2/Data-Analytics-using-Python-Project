```python
def calculate_attendance(attended,total):
  percentage =(attended/total)*100

  if percentage >=75:
    status ="ELIGIBLE"
  else:
    status ="NOT ELIGIBLE"

  return percentage,status

  try:
    total = int(input("Enter total number of classes:"))
    attended = int(input("Enter number of classes attended:"))

    percentage,status = calculate_attendance(attended,total)

    print("Attendance percentage:{percentage:.2f}%")
    print(f"Status:{status}")

  except ZeroDivisionError:
    print("Total classes cannot be zero.")

  except ValueError:
    print(" Please enter valid numbers.")
```
