```python
Sales=[10000,18000, 90000,68000,99000]

total=sum(Sales)

Avg=total/len(Sales)

high_Sales=max(Sales)

low_Sales=min(Sales)

print("Total:", total)

print("Average:", Avg)

print("high_Sales:", high_Sales)

print("low_Sales:", low_Sales)
```

```python
scores = [58,45,73,89,99]

total = sum(scores)
average = total/len(scores)
highest = max(scores)
lowest = min(scores)

print("students scores:", scores)

print("Total marks:", total)

print("average marks:", average)

print("Highest marks:", highest)

print("lowest marks:", lowest)
```
